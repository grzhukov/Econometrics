```python
# 5) пункт 1 задачи.
# HC0 - heteroscedasticity
# сначала в предположении гомоскедастичности, затем гетероскедастичности
from statsmodels.regression.linear_model import OLS
Y = c['price']
X = c['house']
X = sm.add_constant(X)
model = OLS(Y, X)
result_1 = model.fit()

print('1_Homoscedasticity', result_1.summary())
```

```python
Y = c['log_price']
X = c['house']
X = sm.add_constant(X)
model = OLS(Y, X)
result_2 = model.fit(cov_type = 'HC0')
print('2_Heteroscedacity', result_2.summary())
```

```python
# 7-ой пункт 1 задачи.
# 7_1
Y = c['log_price']
X = c['house']
X = sm.add_constant(X)
model = OLS(Y, X)
result_7_1 = model.fit(cov_type = 'HC0')

print('7_1_log_price от house', result_7_1.summary())
```

```python
# 7_2
Y = c['price']
X = c['log_house']
X = sm.add_constant(X)
model = OLS(Y, X)
result_7_2 = model.fit(cov_type = 'HC0')

print('7_2_price от log_house', result_7_2.summary())
```

```python
# 7_3
Y = c['log_price']
X = c['log_house']
X = sm.add_constant(X)
model = OLS(Y, X)
result_7_3 = model.fit(cov_type = 'HC0')

print('7_3_log_price от log_house', result_7_3.summary())
```

Скорректированный (adjusted) $R^2$ равен

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm

%matplotlib inline

# 1-ый пункт 2-ой задачи.
with open(r'C:\Users\grzhu\Desktop\NES\1 module\Python\python-2021-hw2\expend.csv', encoding='utf-8') as f:
    d = pd.read_csv(f)
```

```python
# 2-ой пункт 2-ой задачи.
d[['tot3', 'f3']].describe()
```

```python
# 3-ий пункт 2-ой задачи.
d['log_tot3'] = np.log(d['tot3'])
d['log_f3'] = np.log(d['f3'])
d[['log_tot3', 'tot3', 'log_f3', 'f3']].describe()
```

```python
# 4-ый пункт 2-ой задачи.
from statsmodels.regression.linear_model import OLS
Y = d['log_f3']
X = d['log_tot3']
X = sm.add_constant(X)
model = OLS(Y, X)
result = model.fit(cov_type='HC0')
print(result.t_test('log_tot3 ')) 
```

<font color='red'> Где таблица и интерпретация? -8</font>

```python
# 5-ый пункт 2-ой задачи.
result.t_test('log_tot3 = 0.5')
```

<font color='red'>  Комментарии? -7 </font>

```python
display(result.conf_int(alpha = 0.05))
display(result.conf_int(alpha = 0.1))
```

```python
if result.t_test('log_tot3 = 0.5').pvalue > 0.05:
    print('H_0 не отвергаем на 5%-м уровне')
else:
    print('H_0 отвергаем на 5%-м уровне')
    
if result.t_test('log_tot3 = 0.5').pvalue > 0.1:
    print('H_0 не отвергаем на 10%-м уровне')
else:
    print('H_0 отвергаем на 10%-м уровне')
```
